```python
from dropout_diffusion.gaussian_diffusion import get_named_beta_schedule
```

```python
num_diffusion = 50
```

```python
linear_noise = get_named_beta_schedule("linear", num_diffusion)
```

```python
import numpy as np
```

```python
def alphas_cumprod(betas):
    alphas = 1.0 - betas
    return np.cumprod(alphas, axis=0)
```

```python
linear_alpha = alphas_cumprod(linear_noise)
cosine_alpha = alphas_cumprod(cosine_noise)
exponential_alpha = alphas_cumprod(exponential_noise)
```

```python
plt.plot(linear_alpha, label="linear")
plt.plot(cosine_alpha, label="cosine")
plt.plot(exponential_alpha, label="exponential")
plt.legend()
```
